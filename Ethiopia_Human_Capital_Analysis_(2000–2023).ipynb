# 🇪🇹 Ethiopia Human Capital Analysis (2000–2023)
**Tools:** Python · Pandas · Matplotlib  
**Data:** World Bank World Development Indicators

```python
# Upload your CSV file
from google.colab import files
import pandas as pd

uploaded = files.upload()  # click the button and select your CSV

df = pd.read_csv('Ethiopia_Human_Capital_Master.csv')
print("✅ Data loaded:", df.shape)
df.head()
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()

ax1.bar(df['Year'], df['Mortality rate, under-5 (per 1,000 live births)'],
        color='steelblue', alpha=0.5, label='Under-5 Mortality (per 1,000)')

ax2.plot(df['Year'], df['Prevalence of stunting, height for age (% of children under 5)'],
         color='tomato', linewidth=2.5, marker='o', markersize=4, label='Stunting % (under 5)')

ax1.set_title('Child Health in Ethiopia (2000–2023)', fontsize=14, fontweight='bold', pad=15)
ax1.set_xlabel('Year')
ax1.set_ylabel('Under-5 Mortality Rate (per 1,000)', color='steelblue')
ax2.set_ylabel('Stunting Prevalence (%)', color='tomato')
ax1.set_ylim(0, 200)

lines1, labs1 = ax1.get_legend_handles_labels()
lines2, labs2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labs1 + labs2, loc='upper right', fontsize=9)
ax1.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.savefig('chart1_child_health.png', dpi=150)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()

ax1.plot(df['Year'], df['Primary Enrollment (Female, % gross)'],
         color='royalblue', linewidth=2, marker='o', markersize=4, label='Primary Enrollment Female (%)')
ax1.plot(df['Year'], df['Primary Enrollment (Male, % gross)'],
         color='seagreen', linewidth=2, marker='s', markersize=4, label='Primary Enrollment Male (%)')
ax1.plot(df['Year'], df['School enrollment, secondary, female (% gross)'],
         color='purple', linewidth=2, marker='^', markersize=4, label='Secondary Enrollment Female (%)')

ax2.plot(df['Year'], df['GPI (Female/Male)'],
         color='orange', linewidth=2, linestyle='--', label='Gender Parity Index')

ax1.set_title('Education Indicators in Ethiopia (2000–2023)', fontsize=14, fontweight='bold', pad=15)
ax1.set_xlabel('Year')
ax1.set_ylabel('Enrollment (%)')
ax2.set_ylabel('Gender Parity Index (GPI)', color='orange')

lines1, labs1 = ax1.get_legend_handles_labels()
lines2, labs2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labs1 + labs2, loc='upper left', fontsize=9)
ax1.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.savefig('chart2_education.png', dpi=150)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))
ax2 = ax1.twinx()

ax1.bar(df['Year'], df['Maternal mortality ratio (modeled estimate, per 100,000 live births)'],
        color='salmon', alpha=0.6, label='Maternal Mortality (per 100,000)')

ax2.plot(df['Year'], df['Fertility rate, total (births per woman)'],
         color='navy', linewidth=2.5, marker='o', markersize=4, label='Fertility Rate (births/woman)')

ax1.set_title('Demographics in Ethiopia (2000–2023)', fontsize=14, fontweight='bold', pad=15)
ax1.set_xlabel('Year')
ax1.set_ylabel('Maternal Mortality (per 100,000)', color='salmon')
ax2.set_ylabel('Fertility Rate', color='navy')

lines1, labs1 = ax1.get_legend_handles_labels()
lines2, labs2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labs1 + labs2, loc='upper right', fontsize=9)
ax1.grid(axis='y', linestyle='--', alpha=0.4)

plt.tight_layout()
plt.savefig('chart3_demographics.png', dpi=150)
plt.show()
```
